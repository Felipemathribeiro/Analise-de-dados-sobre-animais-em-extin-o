```python
Animais = pd.read_csv('AnimaisTotais.csv', sep=';')
```

```python
Animais.head(7)
```

```python
p = Animais.sort_values(by= ["Quantidade",],ascending=False).head(10).plot(kind="barh", grid=(True), xticks=(5,10,20,30,40,50,60,70,80,100,110), x=("Estado (2018)"),figsize=(20,10), fontsize=(16))
p.set_xlabel("Quantidade de Animais", fontsize=(16));
p.set_title("Estado de todos os 174 animais que entraram na lista de extinção", fontsize=(20));
p.set_ylabel("Estado (2018)", fontsize=(16));
```

```python
Animais['Estado (2018)'].value_counts().plot(kind='pie', subplots=True, label="", title="Estado dos 174 animais que entrara na lista de ameaçados de extinção na fauna brasileira (2018)" ,figsize=(7, 7))
```
